```python
# 1. Create visualizations to represent the distribution of content over different genres.
genre_counts = net['listed_in'].str.split(', ').explode().value_counts()

plt.figure(figsize=(12, 8))
sns.barplot(x=genre_counts.index, y=genre_counts.values,color = 'red' ,alpha=0.8)

plt.ylabel('Number of Titles', fontsize=12)
plt.xlabel('Genre', fontsize=12)
plt.title('Distribution of Content Over Different Genres', fontsize=14)

plt.xticks(rotation=45, ha='right')

for i, value in enumerate(genre_counts.values):
    plt.text(i, value, str(value), ha='center', va='bottom',color = 'g' ,fontsize=10)

plt.show()
```

```python
# 2. Visualize the distribution of content across release years.
net['release_year'] = pd.to_datetime(net['date_added'], errors='coerce').dt.year

year_counts = net.groupby('release_year').size()

plt.figure(figsize=(12, 8))
sns.barplot(x=year_counts.index, y=year_counts.values,color = 'cyan' ,alpha=0.8)

plt.ylabel('Number of Titles', fontsize=12)
plt.xlabel('Release Year', fontsize=12)
plt.title('Distribution of Content Across Release Years', fontsize=14)

plt.xticks(rotation=45, ha='right')

for i, value in enumerate(year_counts.values):
    plt.text(i, value, str(value), ha='center', va='bottom',color = 'g' ,fontsize=10)

plt.show()
```

```python
# 3. Explore the geographical distribution of content
country_counts = net.groupby('country').size().to_frame(name='count').reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='country', y='count', data=country_counts.sort_values(by='count', ascending=False).head(10),color = 'orange' ,alpha=0.8)

plt.ylabel('Number of Titles', fontsize=12)
plt.xlabel('Country', fontsize=12)
plt.title('Top 10 Countries with Most Content', fontsize=14)

plt.xticks(rotation=45, ha='right')

for i, value in enumerate(country_counts.head(10)['count']):
    plt.text(i, value, str(value), ha='center', va='bottom',color = 'g' ,fontsize=10)

plt.show()
```

```python
# 4.  If there's a temporal component, perform time series analysis to identify trends and patterns over time.

net['release_year'] = pd.to_datetime(net['date_added'], errors='coerce').dt.year

grouped_data = net.groupby('release_year')['title'].count().reset_index(name='count')

plt.figure(figsize=(12, 8))
sns.lineplot(x='release_year', y='count', data=grouped_data, color='purple', marker='o', alpha=0.8)

plt.ylabel('Number of Titles', fontsize=12)
plt.xlabel('Release Year', fontsize=12)
plt.title('Time Series Analysis of Content Release', fontsize=14)

plt.xticks(rotation=45, ha='right')

for i, value in enumerate(grouped_data['count'][::5]):
    plt.text(grouped_data['release_year'][i*5], value, str(value), ha='center', va='bottom', color='g', fontsize=10)

plt.show()
```

```python
# 5. Analyze the distribution of content ratings.
rating_counts = net['rating'].value_counts()

plt.figure(figsize=(12, 8))
sns.barplot(x=rating_counts.index, y=rating_counts.values, color='magenta', alpha=0.8)

plt.ylabel('Number of Titles', fontsize=12)
plt.xlabel('Rating', fontsize=12)
plt.title('Distribution of Content Ratings', fontsize=14)

plt.xticks(rotation=45, ha='right')

for i, value in enumerate(rating_counts.values):
    plt.text(i, value, str(value), ha='center', va='bottom', color='g', fontsize=10)

plt.show()
```

```python
# 6.  Explore the length of movies or episodes and identify any trends.
def categorize_duration(duration):
  """
  Categorize the duration into bins for easier analysis.
  """
  if pd.notna(duration):
    if 'min' in duration:
      minutes = int(duration.split(' ')[0])
      if minutes < 45:
        return '0-45 mins'
      elif minutes < 90:
        return '45-90 mins'
      else:
        return '90+ mins'
    elif 'Season' in duration:
      return 'TV Show'
    else:
      return 'Movie'
  else:
    return 'No duration specified'

net['duration_category'] = net['duration'].apply(categorize_duration)

duration_counts = net.groupby('duration_category').size().to_frame(name='count').reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='duration_category', y='count', data=duration_counts, color='teal', alpha=0.8)

plt.ylabel('Number of Titles', fontsize=12)
plt.xlabel('Duration Category', fontsize=12)
plt.title('Distribution of Content Duration', fontsize=14)

plt.xticks(rotation=45, ha='right')

for i, value in enumerate(duration_counts['count']):
    plt.text(i, value, str(value), ha='center', va='bottom', color='g', fontsize=10)

plt.show()
```

```python
# 7. Analyze trends in the popularity of different genres over time.
grouped_data = net.groupby(['release_year', 'listed_in']).size().unstack().fillna(0)

top_genres = grouped_data.sum().nlargest(10).index
plt.figure(figsize=(12, 8))
for genre in top_genres:
    plt.plot(grouped_data.index, grouped_data[genre], label=genre)

plt.ylabel('Number of Titles', fontsize=12)
plt.xlabel('Release Year', fontsize=12)
plt.title('Popularity Trends of Top 10 Genres Over Time', fontsize=14)
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.show()

for genre in top_genres:
    pct_change = grouped_data[genre].pct_change() * 100
    print(f"Percentage change for {genre}:")
    print(pct_change)
```

```python
# 8.Further explore the distribution of content across different countries and regions.
import matplotlib.pyplot as plt
country_counts = net.groupby('country').size().sort_values(ascending=False)

top_countries = country_counts[country_counts >= 10]

plt.figure(figsize=(15, 6))
top_countries.plot(kind='line', marker='o')
plt.title('Distribution of Content Across Top Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#corellations between duration and ratings.
import pandas as pd
import matplotlib.pyplot as plt
def duration_to_minutes(duration):
  if pd.notna(duration):
    if 'min' in duration:
      return int(duration.split(' ')[0])
    elif 'Season' in duration:
      return None
    else:
      return None
  else:
    return None

net['duration_minutes'] = net['duration'].apply(duration_to_minutes)

rating_mapping = {
    'TV-Y': 1,
    'TV-Y7': 2,
    'TV-Y7-FV': 3,
    'TV-G': 4,
    'TV-PG': 5,
    'TV-14': 6,
    'TV-MA': 7,
    'G': 8,
    'PG': 9,
    'PG-13': 10,
    'R': 11,
    'NC-17': 12,
    'NR': None,
    'UR': None
}
net['rating_numeric'] = net['rating'].map(rating_mapping)

correlation = net['duration_minutes'].corr(net['rating_numeric'], method='spearman')
print("Correlation between duration and rating:", correlation)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='duration_minutes', y='rating_numeric', data=net)
plt.title('Scatterplot of Duration vs. Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating (numeric scale)')
plt.show()
```

```python
# 9.Investigate potential correlations between variables
net['release_year'] = pd.to_numeric(net['release_year'], errors='coerce')

numerical_net = net.select_dtypes(include=['float', 'int'])
correlations = numerical_net.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Variables', fontsize=14)
plt.show()
```

```python
# 10. Evaluate the diversity of content by analyzing the number of unique genres and categories.
unique_genres = net['listed_in'].str.split(', ').explode().unique()
num_unique_genres = len(unique_genres)

# Calculate the number of unique categories
unique_categories = net['type'].unique()
num_unique_categories = len(unique_categories)

# Print the results
print(f"Number of unique genres: {num_unique_genres}")
print(f"Number of unique categories: {num_unique_categories}")
```

```python
# 11. Explore how the characteristics of content (e.g., duration, ratings) have evolved over the years.
grouped_data = net.groupby('release_year')

average_duration = grouped_data['duration'].apply(lambda x: x.str.extract('(\d+)')[0].astype(float).mean())

rating_counts = grouped_data['rating'].value_counts()

plt.figure(figsize=(12, 8))
plt.plot(average_duration.index, average_duration.values)
plt.ylabel('Average Duration (minutes)', fontsize=12)
plt.xlabel('Release Year', fontsize=12)
plt.title('Evolution of Average Content Duration', fontsize=14)
plt.show()

for rating, count_data in rating_counts.groupby(level=0):
  plt.figure(figsize=(12, 8))
  count_data.plot(kind='bar')
  plt.ylabel('Number of Titles', fontsize=12)
  plt.xlabel('Release Year', fontsize=12)
  plt.title(f'Distribution of "{rating}" Ratings Over Years', fontsize=14)
  plt.show()
```
